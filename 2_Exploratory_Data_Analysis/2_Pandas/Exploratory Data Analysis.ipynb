# 1. Accessing Columns using `df[...]`

## Syntax
```python
df['col_name']            # single column (Series)
df[['col_name']]          # single column (DataFrame)
df[['col1', 'col2']]      # multiple columns
```

## Explanation
- Use when you want to access **one or more columns**.
- The output differs:
  - `df['col']` ‚Üí Pandas Series  
  - `df[['col']]` ‚Üí Pandas DataFrame  

## Examples (Using your dataset)
```python
df['Sales']                      # Series
df[['Sales']]                    # DataFrame
df[['Category', 'Profit']]       # Multiple columns
```

---

# 2. Accessing Rows using Basic Python Slicing (`df[start:end:step]`)

## Syntax
```python
df[start : end]         # rows from start to end-1
df[start : end : step]  # with step value
```

## Explanation
- Works like Python list slicing.
- End index is **exclusive**.
- Only works for **rows**, not columns.

## Examples
```python
df[0:1]            # Row 0
df[2:6]            # Rows 2 to 5
df[0:10:2]         # Every alternate row from 0 to 9
```

---

# 3. Selecting Rows + Columns Together (Chained Indexing)

## Syntax
```python
df[start:end][['col1', 'col2']]
```

## Explanation
- First slice rows ‚Üí then select specific columns.

## Example
```python
df[0:5][['Category', 'Sales']]
```

---

#  4. Label-Based Indexing ‚Äî `.loc[]`

`.loc[]` selects data using **row labels** (index names) and **column names**.

## Syntax
```python
df.loc[row_labels , column_labels]
df.loc[start_label : end_label , col_list]
df.loc[:, col_list]                 # all rows
df.loc[row_label, col_label]        # single value
```

## Explanation
- Uses **labels** (not numeric positions).
- End value in slicing is **inclusive**.
- Most human-readable and recommended for teaching.

## Examples
```python
df.loc[0:3, ['Category', 'Sales']]
df.loc[:, ['Profit']]               # Entire Profit column
df.loc[4, 'Sub-Category']           # Output: Storage
```


#  5. Position-Based Indexing ‚Äî `.iloc[]`

`.iloc[]` selects data using **integer positions**.


## Syntax
```python
df.iloc[row_positions , col_positions]
df.iloc[start:end , start:end]
df.iloc[row_position , col_position]    # single value
```

## Explanation
- Uses **positions** (0,1,2‚Ä¶).
- End value is **exclusive**.
- Ideal when you don‚Äôt know column names.

## Examples
```python
df.iloc[0:3, 0:2]        # First 3 rows, first 2 columns
df.iloc[:, 2]            # Sales column (position 2)
df.iloc[1, 3]            # Profit of "Chairs"
```

---

# Final Summary (When to Use What?)

| Method      | Based On | End Included? | When to Use |
|-------------|----------|----------------|--------------|
| `df[col]`   | Column name | ‚Äî | Select 1 or more columns |
| `.loc[]`    | Labels | ‚úî Yes | Most readable; recommended |
| `.iloc[]`   | Positions | No | Numeric indexing |
| `df[start:end]` | Row slicing | No | Simple row slicing only |

---

# 1. Basic Filtering ‚Äî Using Conditions

## Syntax
```python
df[ df['column'] condition ]
```

## Explanation
- Apply condition inside the brackets `[ ]`
- Returns only rows where the condition is **True**

## Examples
```python
df[df['Sales'] > 500]               # Sales greater than 500
df[df['Profit'] < 0]                # Negative profit rows
df[df['Category'] == 'Furniture']   # Category equals Furniture
```

---

# 2. Filtering with Multiple Conditions (AND, OR)

## Syntax
```python
df[(condition1) & (condition2)]     # AND
df[(condition1) | (condition2)]     # OR
df[~(condition)]                    # NOT
```

## Explanation
- Use **&** ‚Üí both conditions must be True  
- Use **|** ‚Üí at least one must be True  
- Use **~** ‚Üí negation  

> Parentheses **must** be used around each condition.

---

## Examples

### Furniture category **AND** Sales > 500
```python
df[(df['Category'] == 'Furniture') & (df['Sales'] > 500)]
```

### Category is Furniture **OR** Technology
```python
df[(df['Category'] == 'Furniture') | (df['Category'] == 'Technology')]
```

### NOT Office Supplies
```python
df[~(df['Category'] == 'Office Supplies')]
```

---

# 3. Filtering Using `.isin()`

## Syntax
```python
df[df['column'].isin([value1, value2, ...])]
```

## Explanation
- Use to filter rows where column value is **in a list**.

## Examples
```python
df[df['Category'].isin(['Furniture', 'Technology'])]
df[df['Sub-Category'].isin(['Phones', 'Tables'])]
```

---

# 4. Filtering Strings ‚Äî Using `.str` Methods

## Syntax
```python
df[df['column'].str.contains('text')]
df[df['column'].str.startswith('text')]
df[df['column'].str.endswith('text')]
```

## Explanation
- Works only with **string columns**
- Very useful for text search and pattern matching

---

## Examples

### Sub-category contains "Phone"
```python
df[df['Sub-Category'].str.contains('Phone')]
```

### Category starts with "Office"
```python
df[df['Category'].str.startswith('Office')]
```

### Sub-category ends with "s"
```python
df[df['Sub-Category'].str.endswith('s')]
```

---

# 5. Filtering Based on Range (BETWEEN)

## Syntax
```python
df[(df['column'] >= lower) & (df['column'] <= upper)]
```

## Explanation
- Used to extract values **between two numbers**.

## Example  
### Sales between 100 and 500
```python
df[(df['Sales'] >= 100) & (df['Sales'] <= 500)]
```

---

# 6. Filtering with `.query()` (Readable Filtering)

## Syntax
```python
df.query("column condition")
```

## Explanation
- Very readable  
- Conditions written like SQL  

## Examples
```python
df.query("Sales > 500")
df.query("Category == 'Furniture'")
df.query("Profit < 0 and Category == 'Furniture'")
```

# Pandas `groupby()` in Pandas

Grouping in Pandas allows you to **split**, **apply**, and **combine** data ‚Äî also called **Split ‚Üí Apply ‚Üí Combine**.

---

## 1. Why GroupBy?

Use `groupby()` to compute:

- Total sales per category  
- Average profit per sub-category  
- Count of records per group  
- Multiple aggregations (sum, mean, max, min)  

---

##  2. Basic Syntax

```python
df.groupby([<categorical_column(s)>])[[<numerical_column(s)>]].agg_function()
```

**Explanation:**

- `groupby([...])` ‚Üí group based on categorical column(s)  
- `[[...]]` ‚Üí select numerical columns  
- `.agg_function()` ‚Üí define aggregation (sum, mean, count, etc.)

---

## 3. Detailed Syntax of groupby()

```python
DataFrame.groupby(
    by=None,
    axis=0,
    level=None,
    as_index=True,
    sort=True
)
```

| Parameter | Meaning |
|----------|---------|
| `by` | Column(s) to group by |
| `axis` | 0 = group by rows |
| `as_index` | True ‚Üí grouped column becomes index |
| `sort` | Sort group keys |
| `level` | Used for MultiIndex |


---

## 4. Split ‚Üí Apply ‚Üí Combine

1. **Splitting**: divide DataFrame into groups  
2. **Applying**: apply aggregation, transformation, or filtration on each group  
3. **Combining**: merge results into a new DataFrame or Series  

image.png

----

## Notes

- `groupby()` returns a **GroupBy object**; no calculation happens until an aggregation is applied  
- Use **`.agg()`** for multiple functions  
- Use **`.transform()`** to return data with original index  
- Use **`.filter()`** to keep/discard groups based on a condition  
- Use **`.apply()`** for custom user-defined functions
