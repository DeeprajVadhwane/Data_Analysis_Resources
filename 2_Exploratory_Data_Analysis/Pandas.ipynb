# **Exploring Data with Pandas**

When you first load your data into a Pandas DataFrame, it's important to take a look at what you're working with. Pandas provides several useful methods to explore and understand your dataset quickly. Let‚Äôs dive into some of these methods that will help you get a good feel for your data:

---
## **1. Exploring the Data: `.head()`, `.tail()`, `.info()`, `.describe()`, `.isnull()`, `.duplicated()`**

Use **`.head()`** to get a quick snapshot of your data. It helps you understand the structure and gives you a preview of the content.

The `.info()` method is like a health check for your DataFrame. It gives you metadata such as the number of entries (rows), column names, data types, and how many non-null values are in each column.

## Non Visual Analysis

Non-visual analysis refers to statistical techniques that allow you to explore your data and extract meaningful insights without plotting graphs or charts. In Pandas, you can easily compute several important statistical measures for numerical and categorical data.

---

### **Numerical Non-Visual Analysis**

For numerical data, you can use a variety of methods to analyze the distribution, spread, and central tendency of your data:

1. **min()**  
   Returns the minimum value in the dataset.
   - **Use Case:** To identify the smallest value in a column.

2. **max()**  
   Returns the maximum value in the dataset.
   - **Use Case:** To identify the largest value in a column.

3. **mean()**  
   Computes the average (mean) of the dataset.
   - **Use Case:** To understand the central tendency.

4. **median()**  
   Finds the middle value of the dataset.
   - **Use Case:** Useful when data is skewed or has outliers.

5. **var()**  
   Calculates the variance, showing the spread of the data.
   - **Use Case:** To understand how much the data deviates from the mean.

6. **std()**  
   Computes the standard deviation, the square root of variance.
   - **Use Case:** To understand the extent of deviation from the mean.

7. **skew()**  
   Measures the skewness (asymmetry) of the data distribution.
   - **Use Case:** To check if your data is skewed left or right.

8. **quantile()**  
   Returns the value below which a given percentage of observations fall.
   - **Use Case:** To calculate percentiles for your data.

9. **kurt()**  
   Measures the "tailedness" of the data distribution.
   - **Use Case:** To understand if the data has heavy tails or outliers.

---

### **Discrete Non-Visual Analysis (Categorical Data)**

For categorical (discrete) data, you can use the following methods to explore the frequency of categories and determine the mode:

1. **count()**  
   Returns the number of non-null entries in a column.
   - **Use Case:** To see how many entries are there in a particular column (excluding null values).

2. **mode()**  
   Identifies the most frequent category in the column.
   - **Use Case:** To find the most common value in a categorical column.

# **Accessing Data in Pandas: `.loc[]` and `.iloc[]`**

When working with data in Pandas, it's essential to know how to access and manipulate specific rows and columns in a DataFrame. Two of the most commonly used methods for this are `.loc[]` and `.iloc[]`. Let‚Äôs dive into the differences between these two and when to use each.

---

## **1. `.loc[]` - Label-based Indexing**

The `.loc[]` method is used for **label-based indexing**, which means you use the row and column labels (names) to select the data.

### **Key Points:**
- You specify the **row labels** and **column labels** inside `.loc[]`.
- Both rows and columns are referenced by their labels, which can be strings or integers if they represent index names.

### **Syntax:**
```python
df.loc[<row_label>, <column_label>]


    df.loc[<start:stop:step>,<start:stop:step>]



````python
df.loc[0:10:2,"Age"]
````

````python
df
````

````python
df.loc[::,"Pclass"]
````

## **2. `.iloc[]` - Position-based Indexing**

The `.iloc[]` method is used for **position-based indexing**, which means you access rows and columns using their integer positions in the DataFrame.

### **Key Points:**
- You specify the **integer positions** (index numbers) for both rows and columns inside `.iloc[]`.
- The index is 0-based, meaning that the first row or column is at position 0, the second at position 1, and so on.

### **Syntax:**
```python
df.iloc[<row_index>, <column_index>]


## **Key Differences between `.loc[]` and `.iloc[]`:**

| **Feature**          | **.loc[ ]**                            | **.iloc[ ]**                            |
|----------------------|---------------------------------------|----------------------------------------|
| **Type of Indexing**  | Label-based (labels of rows/columns) | Position-based (integer index positions) |
| **Accessing Rows**    | Uses the row labels (names)          | Uses integer positions (starting from 0 ) |
| **Accessing Columns** | Uses column labels (names)           | Uses integer positions (starting from 0 ) |
| **Slicing**           | Includes the last element in slice (inclusive) | Does not include the last element (exclusive) |
| **Use Case**          | When you know the row/column labels you want to access | When you know the row/column positions you want to access |

## 1. Merge in Pandas
Merging combines data from multiple DataFrames based on common columns or indices. Similar to SQL joins.




````python
import pandas as pd

# Create sample DataFrames
df1 = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['Mili','Praveen','Harsha']})
df2 = pd.DataFrame({'ID': [2, 3, 4], 'Age': [25, 30, 35]})
````

````python
df1
````

````python
df2
````

````python
df1
````

````python
df2
````

````python
pd.merge(df1,df2,on="ID",how=)
````

````python
# Inner Join
inner_result = pd.merge(df1, df2, on='ID', how='inner')
print('Inner Join:\n', inner_result)
````

````python

# Outer Join
outer_result = pd.merge(df1, df2, on='ID', how='outer')
print('\nOuter Join:\n', outer_result)
````

````python
# Left Join
left_result = pd.merge(df1, df2, on='ID', how='left')
print('\nLeft Join:\n', left_result)

````

````python
# Right Join
right_result = pd.merge(df1, df2, on='ID', how='right')
print('\nRight Join:\n', right_result)
````

## 2. Concat in Pandas
Concatenation combines DataFrames either vertically (default) or horizontally.

## Vertical Stacking (Row-wise)
When stacking vertically, rows from each DataFrame are appended one after another. The number of columns in the DataFrames must match.

````python
# Create sample DataFrames
df1 = pd.DataFrame({'A': ['A0', 'A1'], 'B': ['B0', 'B1']})
df2 = pd.DataFrame({'A': ['A2', 'A3'], 'B': ['B2', 'B3']})
````

````python
df1
````

````python
df2
````

````python
# Vertical Concatenation
vertical_result = pd.concat([df1, df2], axis=0)
print('Vertical Concatenation:\n', vertical_result)

````

````python
# Horizontal Concatenation
horizontal_result = pd.concat([df1, df2], axis=1)
print('\nHorizontal Concatenation:\n', horizontal_result)
````

## 3. Horizontal Stacking (Column-wise)
When stacking horizontally, columns from each DataFrame are combined side-by-side. The number of rows in the DataFrames must match..



````python
# Create sample DataFrames
df1 = pd.DataFrame({'A': ['A0', 'A1'], 'B': ['B0', 'B1']})
df2 = pd.DataFrame({'C': ['C0', 'C1'], 'D': ['D0', 'D1']})

# Horizontal stacking
horizontal_result = pd.concat([df1, df2], axis=1)
print(horizontal_result)
````

The **`groupby()`** method in Pandas is used to group data based on one or more columns, and then apply a function to each group independently. This method is a key part of the **split-apply-combine** pattern, which involves the following steps:

1. **Split**: Split the data into groups based on the values of one or more columns.
2. **Apply**: Apply a function (e.g., mean, sum, count) to each group.
3. **Combine**: Combine the results back into a new data structure (e.g., DataFrame or Series).

##üìå What is a Pivot Table?
A pivot table is a powerful tool in pandas used to summarize, aggregate, and analyze data. It allows you to reshape and group data by specifying one or more columns as the index, columns, and values.

````python
import pandas as pd

# Sample retail sales data
data = {
    'Date': ['2024-01-01', '2024-01-01', '2024-01-02', '2024-01-02', '2024-01-03', '2024-01-03'],
    'Region': ['North', 'South', 'North', 'South', 'East', 'East'],
    'Product': ['Laptop', 'Laptop', 'Phone', 'Phone', 'Laptop', 'Phone'],
    'Category': ['Electronics', 'Electronics', 'Electronics', 'Electronics', 'Electronics', 'Electronics'],
    'Sales': [1000, 1200, 800, 900, 1100, 950],
    'Quantity': [5, 6, 8, 10, 4, 5]
}

df = pd.DataFrame(data)
print(df)
````

### 1. Total Sales by Region and Product:

````python
import pandas as pd
pd.pivot_table(df,index="Region",
               values="Sales",
               aggfunc='sum')

````

````python
pd.pivot_table(df,index="Region",
               columns="Product",
               values="Sales",
               aggfunc='sum')
````

````python
df.groupby('Region')['Sales'].sum()
````

````python
pivot_sales = pd.pivot_table(df,
                             index='Product',
                             columns='Region',
                             values='Sales',
                             aggfunc='sum')
print(pivot_sales)
````

### 2. Average Quantity Sold per Product:

````python
pivot_quantity = pd.pivot_table(df,
                                index='Product',
                                values='Quantity',
                                aggfunc='mean')
print(pivot_quantity)
````

### 3. Multiple Aggregations: Sales and Quantity:

````python
pivot_multi = pd.pivot_table(df,
                             index='Region',
                             values=['Sales', 'Quantity'],
                             aggfunc={'Sales': 'sum', 'Quantity': 'sum'})
print(pivot_multi)
````

### 4. Adding Totals (Margins):

````python
pivot_margins = pd.pivot_table(df,
                               index='Region',
                               columns='Product',
                               values='Sales',
                               aggfunc='sum',
                               margins=True,
                               margins_name='Total')
print(pivot_margins)
````

### 5. Handling Missing Data (fill_value):

````python
pivot_fill = pd.pivot_table(df,
                            index='Product',
                            columns='Region',
                            values='Sales',
                            aggfunc='sum',
                            fill_value=0)
print(pivot_fill)
````

````python
pivot_fill.plot(kind='bar')
````
