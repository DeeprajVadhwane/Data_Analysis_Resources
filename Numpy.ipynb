# Understanding Arrays in Programming

### What is an Array?
- A structure for storing and retrieving data.
- Visualized as a grid, where each cell holds one element of the data.

#### Types of NdArrays
1. **One-Dimensional Array**  
   - Represents data in a linear structure (like a list).  
   - **Example**: `[1, 2, 3, 4, 5]`

2. **Two-Dimensional Array**  
   - Organized as rows and columns, similar to a table.  
   - **Example**:  
     ```
     [[1, 2, 3],
      [4, 5, 6],
      [7, 8, 9]]
     ```

3. **Three-Dimensional Array**  
   - Represents a collection of tables stacked together, like pages.  
   - **Example**:  
     ```
     [
      [[1, 2], [3, 4]],
      [[5, 6], [7, 8]]
     ]
     ```

4. **N-Dimensional Array**  
   - Generalized concept in NumPy as the `ndarray` class for any number of dimensions.

### Restrictions in NumPy Arrays
1. **Data Type Consistency**  
   - All elements must be of the same data type.  
   - **Example**: `[1, 2, 3]` (valid) vs. `[1, 'two', 3]` (invalid).

2. **Fixed Size**  
   - The total size of the array cannot be altered after creation.  
   - **Example**: Cannot add or remove elements directly like Python lists.

3. **Rectangular Shape**  
   - Arrays must be uniform in shape (no jagged edges).  
   - **Example**:  
     Valid:  
     ```
     [[1, 2],
      [3, 4]]
     ```  
     Invalid:  
     ```
     [[1, 2],
      [3]]
     ```

### Advantages of NumPy Arrays
1. **Performance**  
   - Faster operations due to uniformity and optimized implementations.  

2. **Memory Efficiency**  
   - Requires less memory compared to other data structures.  

3. **Ease of Use**  
   - Simplifies operations on large datasets using high-level functions.  

> **Note**: For the rest of this document, "array" refers to a NumPy `ndarray`.
